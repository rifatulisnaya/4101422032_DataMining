## Identifikasi Outliers

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Kolom numerik yang ingin dicek outliernya
num_cols = ['PassengerId','Survived','Pclass','Age','SibSp','Parch','Fare']

# Visualisasi Boxplot
plt.figure(figsize=(15, 8))
for i, col in enumerate(num_cols, 1):
    plt.subplot(3, 4, i)
    sns.boxplot(y=data[col])
    plt.title(col)
plt.tight_layout()
plt.show()
```

```python
def detect_outliers_iqr(data, column):
    Q1 = data[column].quantile(0.25)  # Kuartil 1
    Q3 = data[column].quantile(0.75)  # Kuartil 3
    IQR = Q3 - Q1  # Rentang interkuartil
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    return data[(data[column] < lower_bound) | (data[column] > upper_bound)]

# Cek jumlah outlier untuk setiap kolom numerik
for col in num_cols:
    outliers = detect_outliers_iqr(data, col)
    print(f"{col}: {len(outliers)} outliers")
```

Terdapat outliers pada dataset tersebut yaitu pada kolom Age, SibSp, Parch, dan Fare.

**Outlier** adalah data yang memiliki nilai sangat berbeda atau menyimpang secara ekstrem dibandingkan dengan data lainnya dalam satu dataset.

Untuk menangani outlier tersebut, akan diterapkan metode Winsorizing, yakni mengganti nilai outlier dengan nilai batas kuartil. Hal ini dilakukan karena kolom-kolom yang mengandung outlier merupakan bagian data yang dianggap penting.

```python
def winsorize_iqr(data, column):
    Q1 = data[column].quantile(0.25)
    Q3 = data[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    data[column] = np.where(data[column] < lower_bound, lower_bound, data[column])
    data[column] = np.where(data[column] > upper_bound, upper_bound, data[column])

# Terapkan Winsorizing
for col in num_cols:
    winsorize_iqr(data, col)
```

# 3. Modelling

Data training merupakan bagian dari dataset yang dimanfaatkan untuk melatih model dalam proses *machine learning*. Pada tahap ini, model mempelajari pola dan hubungan antara fitur (variabel independen) dengan label atau target (variabel dependen). Melalui data training, model menyesuaikan parameter internalnya agar mampu menghasilkan prediksi yang akurat terhadap data baru.

Di sisi lain, data testing atau data uji adalah bagian dataset yang digunakan untuk mengevaluasi performa model setelah proses pelatihan selesai. Data ini tidak dilibatkan dalam pelatihan, sehingga berguna untuk menguji sejauh mana model mampu menggeneralisasi terhadap data baru yang belum pernah ditemui sebelumnya. Dengan data testing, kita bisa menilai kemampuan model dalam memprediksi label secara tepat, sekaligus memastikan bahwa model tidak sekadar menghafal data training, melainkan benar-benar memahami pola umum.

**Alasan perlu membagi data training dan data testing.**

Pemisahan data menjadi data training dan data testing penting dilakukan untuk memastikan bahwa model memiliki kemampuan generalisasi yang baik. Jika seluruh dataset hanya digunakan untuk melatih model, maka ada risiko *overfitting*, yaitu kondisi di mana model terlalu fokus pada data pelatihan dan gagal melakukan prediksi akurat terhadap data baru. Dengan adanya data testing, evaluasi kinerja model dapat dilakukan secara lebih objektif dan adil.

Selain itu, pembagian ini mensimulasikan kondisi penggunaan model di dunia nyata, di mana model dihadapkan pada data yang belum pernah dilihat sebelumnya. Model yang mampu mempertahankan kinerjanya pada data testing menunjukkan bahwa pola yang dipelajari dari data training memang merepresentasikan karakteristik data secara umum. Oleh sebab itu, pembagian data menjadi training dan testing sangatlah penting untuk memastikan prediksi model dapat diandalkan dalam praktik sebenarnya.

## Algoritma Naive Bayes

---

**Naive Bayes** merupakan algoritma klasifikasi yang berlandaskan teorema Bayes dengan asumsi bahwa setiap fitur bersifat independen satu sama lain. Dengan kata lain, meskipun dalam kenyataannya fitur-fitur bisa saling berhubungan, algoritma ini menganggap masing-masing fitur tidak saling bergantung.

**Alur kerja Naive Bayes adalah sebagai berikut:**

1. **Probabilitas Prior:** Menentukan probabilitas awal dari masing-masing kelas target (contohnya pada kolom Survived, nilai "0 = Selamat" atau "1 = Tidak Selamat").
2. **Likelihood:** Menghitung peluang munculnya nilai fitur tertentu dalam masing-masing kelas.
3. **Posterior:** Menggunakan teorema Bayes untuk menghitung probabilitas suatu kelas berdasarkan fitur-fitur yang diamati.
4. **Klasifikasi:** Memilih kelas dengan nilai probabilitas posterior tertinggi sebagai hasil prediksi.

## Algoritma Neural Network

---



**Neural Networks (Jaringan Saraf Tiruan)** adalah model pembelajaran yang kompleks, terinspirasi dari cara kerja otak manusia. Model ini terdiri dari beberapa lapisan (*layers*) yang memproses data secara bertahap untuk menghasilkan prediksi.

Berikut adalah mekanisme kerja Neural Networks:

1. **Input Layer:** Data berupa fitur-fitur seperti Age, Fare, dan lainnya dimasukkan ke dalam input layer, di mana setiap fitur diwakili oleh satu neuron.

2. **Hidden Layers:** Data kemudian diteruskan ke satu atau lebih hidden layers. Pada tahap ini, setiap neuron mengalikan inputnya dengan bobot tertentu, menerapkan fungsi aktivasi seperti ReLU atau sigmoid, lalu meneruskan hasilnya ke neuron di lapisan berikutnya. Hidden layers memungkinkan jaringan untuk memahami pola-pola non-linear dalam data.

3. **Output Layer:** Di lapisan output, jaringan menghasilkan prediksi akhir berdasarkan informasi yang diproses di hidden layers. Untuk tugas klasifikasi biner, output biasanya berupa probabilitas yang mewakili kemungkinan termasuk dalam suatu kelas.

4. **Backpropagation:** Setelah prediksi dihasilkan, model menghitung error dengan membandingkannya terhadap label sebenarnya. Kemudian, melalui proses backpropagation, bobot-bobot di dalam jaringan diperbarui untuk meminimalkan error tersebut. Proses ini dilakukan berulang-ulang (melalui banyak epoch) guna meningkatkan kinerja model.

```python
#Menguji parameter yang paling optimum untuk model

from sklearn.neural_network import MLPClassifier
from sklearn.model_selection import GridSearchCV

from sklearn.model_selection import StratifiedKFold

cv_strategy = StratifiedKFold(n_splits=5, shuffle=True, random_state=100)



# Tentukan parameter yang ingin diuji
parameters = {
    'hidden_layer_sizes': [(50,), (100,), (50, 30), (100, 50)],
    'learning_rate_init': [0.001, 0.01, 0.1],
    'max_iter': [100, 200, 300]
}

# Buat model MLPClassifier
clf = MLPClassifier(random_state=100)

# Inisialisasi GridSearchCV
grid_search = GridSearchCV(clf, parameters, cv=cv_strategy)
# cv merupakan jumlah lipatan validasi silang (cross-validation)

# Lakukan pencarian grid pada data training
grid_search.fit(X, y)
```

## Algoritma K-Nearest Neighbor

---

**K-Nearest Neighbors (KNN)** adalah algoritma klasifikasi berbasis instance yang menentukan prediksi dengan cara mencari data-data terdekat di sekitar data yang ingin diklasifikasikan. KNN mengukur kedekatan ini menggunakan jarak, seperti jarak Euclidean.

Berikut langkah-langkah kerja KNN:

1. **Menentukan Nilai K:** Tetapkan jumlah tetangga terdekat (k) yang akan digunakan. Sebagai contoh, jika k=3, maka prediksi dibuat berdasarkan 3 data terdekat dari data uji.
2. **Menghitung Jarak:** Ukur jarak antara data uji dengan semua data di dataset pelatihan. Umumnya, jarak yang digunakan adalah Euclidean Distance.
3. **Menentukan Kelas:** Pilih k data terdekat, lalu tentukan kelas berdasarkan mayoritas dari tetangga tersebut. Misalnya, jika mayoritas tetangga termasuk dalam kategori "Selamat", maka data uji akan diprediksi sebagai "Selamat".

```python
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.metrics import confusion_matrix

def plot_confusion_matrix(y_true, y_pred, model_name):
  """
  Membuat visualisasi confusion matrix.

  Args:
    y_true: Nilai sebenarnya.
    y_pred: Nilai prediksi.
    model_name: Nama model.
  """
  cm = confusion_matrix(y_true, y_pred)
  plt.figure(figsize=(6, 6))  # Mengatur ukuran gambar
  sns.heatmap(cm, annot=True, fmt="d", cmap="Blues",
              xticklabels=['Tidak Selamat', 'Selamat'],
              yticklabels=['Tidak Selamat', 'Selamat'])
  plt.title(f'Confusion Matrix - {model_name}')
  plt.xlabel('Prediksi')
  plt.ylabel('Aktual')
  plt.show()

plot_confusion_matrix(y_test, y_pred_nb, "Naive Bayes")
```
